# Pandas interpolate()

```python
# Pandas interpolate function is basically used to fill NaN value in dataframe or series
```

```python
# it is very powerful function 
# uses various interpolation techniques
```

```python
df=pd.read_csv("D:\CSV\student_results - Copy.csv")
df
```

```python
df.interpolate() # Interpolate function is guessing the missing value in data set.
```

```python
df.interpolate(method = 'linear')# method = 'linear ' is default dataframe
```

```python
df.interpolate(method = 'time')
```
