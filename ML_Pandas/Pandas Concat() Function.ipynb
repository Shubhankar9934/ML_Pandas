# Pandas Concat() Function

```python
# Pnadas provides various facilities for easily combining 
# together series , DataFrame and Panel objects.
```

```python
import pandas as pd 
```

```python
sr1=pd.Series([0,1,2])
sr1
```

```python
sr2 = pd.Series([3,4,5])
sr2
```

```python
pd.concat([sr1,sr2])
# here we concat two equal series 
```

```python
sr3 = pd.Series([3,4,5,6])
sr3
```

```python
pd.concat([sr1,sr3])
# two Unequal series are concat
```

```python
df1=pd.DataFrame({'ID':[1,2,3,4],
             'Name':['A','B','C','D'],
             'Class':[5,6,7,8]})
df1
```

```python
df2=pd.DataFrame({'ID':[5,6,7,8],
             'Name':['E','F','G','H'],
             'Class':[9,10,11,12]})
df2
```

```python
pd.concat([df1,df2])
```

```python
pd.concat([df2,df1])
```

```python
pd.concat([df2,df1],axis = 1)
```

```python
pd.concat([df1,df2], axis = 0) 
```

```python
pd.concat([df1,df2], axis = 0,ignore_index=True) 
# To overcome from the problem of repetetion of same index
#we use ignore_index = True
```

```python
pd.concat([df1,df2], axis = 0,ignore_index=False) 
# default.
```

```python
df1=pd.DataFrame({'ID':[1,2,3,4],
             'Name':['A','B','C','D'],
             'Class':[5,6,7,8]})
df1
```

```python
df2=pd.DataFrame({'ID':[3,4],
             'Name':['C','D'],
             'Class':[7,8]})
df2
```

```python
pd.concat([df1,df2])
```

```python
pd.concat([df1,df2], axis = 1)
```

```python
pd.concat([df1,df2], axis = 1, join = 'inner')
# join = 'inner' takes intersection of both data frame.
```

```python
pd.concat([df1,df2], axis = 1 , join = 'outer')
# join = 'outer' takes Union of Both DataFrame.
```

```python
# join_axes : List of Index objects.
pd.concat([df1,df2], axis = 1 , ignore_index=True)
```

```python
df1=pd.DataFrame({'ID':[1,2,3,4],
             'Name':['A','B','C','D'],
             'Class':[5,6,7,8]})
df1
```

```python
df2=pd.DataFrame({'ID':[5,6,7,8],
             'Name':['E','F','G','H'],
             'Class':[9,10,11,12]})
df2
```

```python
pd.concat([df1,df2],keys = ['df1','df2'])
```

```python
pd.concat([df1,df2],keys = ['First df','Second df'])
```

```python
pd.concat([df1,df2],keys = ['First df','Second df'], axis = 1)
```
