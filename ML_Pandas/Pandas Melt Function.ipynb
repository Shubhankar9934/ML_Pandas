```python
#Pandas melt function is used to transform or reshape data.
#Syntax : pandas.melt()
# Parameters:
#     Frame 
#     id_vars
#     value_name
#     value_vars
#     var_name
```

```python
import pandas as pd
```

```python
df=pd.read_csv("D:\CSV\movies.csv")
df
```

```python
pd.melt(frame=df)
```

```python
pd.melt(frame =df,id_vars =['Genre'])
```

```python
pd.melt(frame =df,id_vars =['Year'],value_vars = ['Genre'])
```

```python
pd.melt(frame =df,id_vars =['Year'],value_vars = ['Rotten Tomatoes %'])
```

```python
pd.melt(frame =df,id_vars =['Year'],value_vars = ['Rotten Tomatoes %'],var_name = 'Catergory')
```

```python
pd.melt(frame =df,id_vars =['Year'],value_vars = ['Rotten Tomatoes %'],var_name = 'Catergory',value_name='Data')
```
