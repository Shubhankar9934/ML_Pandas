# Pandas Groupby Function

```python
# Pandas groupby function is used to split the into groups based on some criteria
```

```python
# Groupby opression 
# 1) Splitting the object
# 2)Applying a function 
# 3) Combining the result
```

```python
import pandas as pd
```

```python
df= pd.read_csv("D:\CSV\student_results.csv")
df
```

```python
gr1=df.groupby(by='Class')
gr1

# Signature:
# df.groupby(
#     by=None,
#     axis: 'Axis' = 0,
#     level: 'Level | None' = None,
#     as_index: 'bool' = True,
#     sort: 'bool' = True,
#     group_keys: 'bool' = True,
#     squeeze: 'bool | lib.NoDefault' = <no_default>,
#     observed: 'bool' = False,
#     dropna: 'bool' = True,
# ) -> 'DataFrameGroupBy'
```

```python
gr1.groups
```

```python
df.groupby(by= ['Student ID','Class','Percantege',]).groups 
# group by multiple column
```
