```python
# Fillina Fills the NaN values with Given Values (input)
#syntax:DataFrame.fillna()
```

```python
import pandas as pd
```

```python
df=pd.read_csv('D:\\CSV\\Fortune_10 - Copy.csv')
df
```

# value


```python
df.fillna(0)
   #value : scalar, dict, Series, or DataFrame
   # Value to use to fill holes (e.g. 0), alternately a
   # dict/Series/DataFrame of values specifying which value to use for
   # each index (for a Series) or column (for a DataFrame).  Values not
   # in the dict/Series/DataFrame will not be filled. This value cannot
   # be a list.
```

```python
df.fillna(2)
```

```python
df
```

```python
df.fillna({'Industry':'none','Inception':0,'Expenses':'0.0','Growth':'0%'})
```

```python
df
```

# method

```python
df.fillna(method='ffill') # This is method for forward fill means filling forward value
```

```python
df
```

```python
df.fillna(method='pad') # works same as a filling  forward value .
```

```python
df
```

```python
df.fillna(method='bfill')# filling backward value.
```

# axis

```python
df.fillna(method= 'ffill',axis = 0) 
#axis : {0 or 'index', 1 or 'columns'}
    #Axis along which to fill missing values.
```

```python
df.fillna(method='ffill', axis = 1)
```

```python
df.fillna(method = 'bfill', axis = 1)
```

```python
df.fillna(limit= 10, value= 0)
```

# limit

```python
df.fillna(method= 'ffill', limit=1) 
```

# inplace

```python
df.fillna(inplace= True, value=5)
df
```
