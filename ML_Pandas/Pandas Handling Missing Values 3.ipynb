```python
import pandas as pd
```

```python
df=pd.read_csv('D:\\CSV\\Fortune_10 - Copy.csv')
df
```

# dropna()

```python
df.dropna() # Remove Missing Values.
```

```python
df.dropna(axis =1) # Drop all the columns having Missing Values.
```

# how='any'

```python
df.dropna(how ='any') # Drop all the rows having Missing Values.
```

# axis

```python
df.dropna(how='any',axis = 1) # Drop all the rows and columns having Missing values
```

# how

```python
df.dropna(how='all')
```

# thresh

```python
df.dropna(thresh=1) # it prints all rows and columns having at least some values with nan but if all values is NaN then it drops that rows and columns.
```

# Subset parameter

```python
# subset Parameter
df.dropna(subset= ['ID','data','Name','Inception'])
```

# inplace

df.dropna(inplace =True) # It Drop all NaN value from data set and update the same data frame
