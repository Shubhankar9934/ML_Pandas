# Pandas Merge Function

```python
# Pandas merge connects coloumns or indexes in dataframe 
# based on one oe more keys 
```

```python
df1 = pd.DataFrame({'ID':[1,2,3,4],
                  'Class':[9,10,11,12]})
df1
```

```python
df2 = pd.DataFrame({'ID':[1,2,3,4],
                   'Name':['shubh','jyoti','khush','Nikku']})
df2
```

```python
df4=pd.merge(df1,df2, on = 'ID')
df4
```

```python
df3=pd.DataFrame({'Section':['A','B','C','D'],
                 'Class':[9,10,11,12]})
df3
```

```python
pd.merge(df4,df3, on = 'Class')
```

```python
df5=pd.DataFrame({'ID':[1,2,3,5],
             'Strength':[30,31,32,33]})
df5
```

```python
df1
```

```python
pd.merge(df1,df5,on="ID") 
# It takes default Value as how=Inner and gives the 
# intersection and not printing unmatched value.
```

```python
pd.merge(df1,df5, on = 'ID', how ='left')
# It gives output as a left data frame.
```

```python
pd.merge(df1,df5, on ='ID', how = 'right')
# It gives output as a Right data frame.
```

```python
pd.merge(df1,df5,on = 'ID', how = 'outer')
# It print all data value and left missing value as NaN.
```

```python
pd.merge(df1,df5,on = 'ID', how = 'outer', indicator = True)
# it gives All idea about which value is taken and which value is left.
#To know the information about each row we use indicatoar parameter. 
```
